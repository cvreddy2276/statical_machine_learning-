```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
d = pd.read_csv('/content/drive/MyDrive/Real estate valuation data set.csv')
```

```python
d
```

```python
x = d.drop('Y house price of unit area', axis=1)
y = d['Y house price of unit area']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7)
```

# Linear Regression

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
def predict(x):
  return model.predict(x)
def errors(y_test, y_pred):
  mse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  return mse, mae
```

```python
test_pred = predict(x_test)
mse_test, mae_test = errors(y_test, test_pred)
print(mse_test, mae_test)
```

```python
train_pred = predict(x_train)
mse_train, mae_train = errors(y_train, train_pred)
print(mse_train, mae_train)
```

# Lasso regression

```python
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(x_train, y_train)
y_pred_lasso = lasso_reg.predict(x_train)
mse_train = mean_squared_error(y_train, y_pred_lasso)
mae_train = mean_absolute_error(y_train, y_pred_lasso)
print(f"  Mean Squared Error: {mse_train}")
print(f"  Mean Absolute Error: {mae_train}")
```

```python
lasso_reg.fit(x_train, y_train)
y_pred_lasso = lasso_reg.predict(x_test)
mse_test = mean_squared_error(y_test, y_pred_lasso)
mae_test = mean_absolute_error(y_test, y_pred_lasso)
print(f"  Mean Squared Error: {mse_test}")
print(f"  Mean Absolute Error: {mae_test}")
```

# Ridge Regression

```python
ridge_reg = Ridge(alpha=0.1)
ridge_reg.fit(x_train, y_train)
y_pred_ridge = lasso_reg.predict(x_train)
mse_train = mean_squared_error(y_train, y_pred_ridge)
mae_train = mean_absolute_error(y_train, y_pred_ridge)
print(f"  Mean Squared Error: {mse_train}")
print(f"  Mean Absolute Error: {mae_train}")
```

```python
ridge_reg.fit(x_test, y_test)
y_pred_ridge = ridge_reg.predict(x_test)
mse_test = mean_squared_error(y_test, y_pred_ridge)
mae_test = mean_absolute_error(y_test, y_pred_ridge)
print(f"  Mean Squared Error: {mse_test}")
print(f"  Mean Absolute Error: {mae_test}")
```

# Support Vector Machine (SVM)

```python
sr = SVR(kernel='linear')
sr.fit(x_train, y_train)
y_pred_sr = sr.predict(x_train)
mse_train = mean_squared_error(y_train, y_pred_sr)
mae_train = mean_absolute_error(y_train, y_pred_sr)
print(f"  Mean Squared Error: {mse_train}")
print(f"  Mean Absolute Error: {mae_train}")
```

```python
sr.fit(x_train, y_train)
y_pred_sr = sr.predict(x_test)
mse_test = mean_squared_error(y_test, y_pred_sr)
mae_test = mean_absolute_error(y_test, y_pred_sr)
print(f"  Mean Squared Error: {mse_test}")
print(f"  Mean Absolute Error: {mae_test}")
```

# K - Nearest Neighbor  (KNN)

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_train)
mse_train = mean_squared_error(y_train, y_pred_knn)
mae_train = mean_absolute_error(y_train, y_pred_knn)
print(f"  Mean Squared Error: {mse_train}")
print(f"  Mean Absolute Error: {mae_train}")
```

```python
from sklearn.utils import resample
df_resampled = resample(df, replace=True, n_samples=len(df), random_state=42)
x_resampled = df_resampled.drop('Y house price of unit area', axis=1)
y_resampled = df_resampled['Y house price of unit area']
x_train_resampled, x_test_resampled, y_train_resampled, y_test_resampled = train_test_split( x_resampled, y_resampled, test_size=0.6, random_state=42)
```

# Linear Regression(Bootstrap)

```python
n_iterations = 100
results = []

for i in range(n_iterations):
    x_train_resampled, y_train_resampled = resample(x_train, y_train, random_state=i)
    model = LinearRegression()
    model.fit(x_train_resampled, y_train_resampled)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    results.append(mse)

print(f"Mean MSE across bootstrap iterations: {np.mean(results)}")
```

# Lasso Regression

```python
n_iterations = 100
results = []

for i in range(n_iterations):
    x_train_resampled, y_train_resampled = resample(x_train, y_train, random_state=i)
    model = Lasso()
    model.fit(x_train_resampled, y_train_resampled)
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    results.append(mse)

print(f"Mean MSE across bootstrap iterations: {np.mean(results)}")
```
